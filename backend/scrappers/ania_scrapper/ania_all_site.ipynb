```python
import requests
from bs4 import BeautifulSoup
```

```python
url = f'https://aniagotuje.pl/sitemap.xml'
headers = {
    "User-Agent": "Mozilla/5.0"
}
page = requests.get(url, headers=headers)
```

```python
soup = BeautifulSoup(page.content,features='xml')
```

```python
def get_links(soup):
    almost_links = soup.find_all("loc")
    results =[]
    for item in almost_links:
        for link in item.children:
            results.append(link)
            #print(link)
    return results
get_links(soup)
```

```python
def save_links_to_file(links):
    with open("links.csv","w") as file:
        for link in links:
            file.write(link+"\n")
```

```python
save_links_to_file(get_links(soup))
```
