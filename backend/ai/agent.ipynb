```python
from ollama import chat
```

```python

response = chat(
    model='hf.co/speakleash/Bielik-11B-v2.2-Instruct-GGUF:latest',
    messages=[{'role': 'user', 'content': 'Jak dobrze znasz polski? Lepiej pracujesz po polsku czy po angielsku? Czy może kombinacją obu?'}],
)
print(response['message']['content'])
```
